```python
import pandas as pd
import rasterio as rio
from rasterio.plot import show
import matplotlib.pyplot as plt

tif_df = pd.read_csv("./bounds.csv", sep="|")
tif_df.head()
```

```python
tif_collection = list(tif_df.File)
tif_names = list(tif_df.Name)
for idx, file in enumerate(tif_collection):
    if 10 < idx < 20:
        plt.figure(figsize=(14,10))
        plt.title(tif_names[idx])
        show(rio.open(file), cmap="CMRmap")
```
